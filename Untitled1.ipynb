```python
import tensorflow as tf
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
base_dir = "E:\\diabetes_disease_classification_FYP\\retina_images"

# Initialize empty lists to store the images and labels
images = []
labels = []

# Loop through each class' folder
for class_label in range(5):
    # Get a list of all the image file names in this folder
    image_files = os.listdir(os.path.join(base_dir, str(class_label)))
    
    # Load each image, convert it to an array, and append it to the images list
    for image_file in image_files:
        image = load_img(os.path.join(base_dir, str(class_label), image_file), target_size=(512, 512))
        images.append(img_to_array(image))
        
        # Append this image's class label to the labels list
        labels.append(class_label)

# Convert the lists to numpy arrays
images = np.array(images)
labels = np.array(labels)
```

```python
# Split the data into training and testing sets
train_images, test_images, train_labels, test_labels = train_test_split(images, labels, test_size=0.2, stratify=labels, random_state=42)

# Further split the training data into training and validation sets
train_images, val_images, train_labels, val_labels = train_test_split(train_images, train_labels, test_size=0.25, stratify=train_labels, random_state=42)
```

```python
def plot_images(images, labels):
    # Create a figure to contain the plot.
    fig = plt.figure(figsize=(10, 10))

    # Loop through each class (from 0 to 4)
    for class_label in range(5):
        # Find the index of the first occurrence of this class label in the labels array
        index = np.where(labels == class_label)[0][0]
        
        # Add a subplot for this class
        ax = fig.add_subplot(2, 3, class_label + 1)
        
        # Display the image
        ax.imshow(images[index].astype('uint8'))
        
        # Set the title of the subplot
        ax.set_title("Class: {}".format(class_label))

    # Show the plot
    plt.show()

# Plot the training images
plot_images(train_images, train_labels)
```

```python
train_labels = to_categorical(train_labels)
val_labels = to_categorical(val_labels)
test_labels = to_categorical(test_labels)

# Initialize the ImageDataGenerators
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Create generators to read images from the numpy arrays
train_generator = train_datagen.flow(train_images, train_labels, batch_size=8)
val_generator = val_datagen.flow(val_images, val_labels, batch_size=8)
test_generator = test_datagen.flow(test_images, batch_size=1)
```

```python
from keras.applications import VGG16
```

```python
vgg16 = VGG16(include_top=False, weights='imagenet', input_shape=(512,512,3))

model = Sequential()
model.add(vgg16)
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))  # Added dropout
model.add(Dense(5, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
history = model.fit(train_generator,
                    steps_per_epoch=15,
                    validation_data=val_generator,
                    epochs=30,
                    validation_steps=5,
                    verbose=1)
```
